## Imports

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from statistics import mean, stdev
```

## Building base model

```python
def regression_model():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(8,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
model = regression_model()
```

```python
report = pd.DataFrame(columns = ['Assignment','Model, data and epoch information', 'MSE'])
```

## Data import and cleaning

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

```python
concrete_data_columns = concrete_data.columns
X = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']]
Y = concrete_data['Strength']
X.head()
```

```python
Y.head()
```

# Part A: Base model

## Part a1: Split data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
```

## Part a2: Training with 50 epochs

```python
model.fit(X_train, Y_train, validation_split=0.3, epochs=50, verbose=2)
```

## Part a3: Evaluation

```python
Y_predicted = predict()
mse = calculate_mse()
report.loc[0] = ['A2', 'Single run, one hidden layer, 50 epochs', round(mse,2)]
print('The mean square error' , round(mse))
```

## Part a4: Repeat 50 times

```python
mses = []
for i in range(50):
    model = regression_model()
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
    model.fit(X_train, Y_train, validation_split=0.3, epochs=50, verbose=0)
    Y_predicted = predict()
    mse = calculate_mse()
    mses.append(mse)
    print('Run', i + 1 , 'has an mse of', round(mse,2))
```

```python
msesmean = mean(mses)
msesstd = stdev(mses)
report.loc[1] = ['A4', 'Avarage of 50 runs, one hidden layer, 50 epochs', round(msesmean)]
print('The mean of mean square errors' , str(round(msesmean,2)))
print('Standard deviation of mean square errors' , str(round(msesstd,2)))
```

# Part D: Increase the number of hidden layers

```python
def l3_regression_model():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(8,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```
